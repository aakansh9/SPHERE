# Step 1)

## Accelerometer data processing

- raw
- smoothened
- frequency filters (filter bank of low pass filters)
- frequency filters (high pass filters)
- integral (speed)
- diff (jerk)

## Cross-axis interactions

- x+y, y+z, x+z, x+y+z
- x-y, y-z, z-x
- event specific custom interaction

## Features from accelerometer data

- stat_feat (mean, sd, quantiles) of each axis (x,y,z) / cross-axis-interaction (x+y, ...) in windows of size s (1 sec, 2 sec, 3 sec, ... ) after processing (raw, smooth, filter bank, jerk, ... )
- Binary features related to existence of specific-event-patterns (threshholds, oscillations, other patterns) in windows of size s.

### Choosing window size
- rare events have more weight in metric and they occur in small windows.
- common events occur in big windows.

```python
def window_feat(t_start, t_end, df, columns):
    
    window_size = t_end - t_start
    window_data = df[(df['t'] >= t_start) & (df['t'] < t_end)]
    window_data = window_data[columns]
    
    F = [] # empty features list
    fnames = [] # empty feature names list
    base_names = [ (column + '_W' + str(window_size)) for column in window_data.columns]
    
    # quantiles
    F = F + window_data.quantile(q = [0, 0.25, 0.50, 0.75, 1]).values.flatten().tolist()
    names = ( names + [ colname + '_0Q' for colname in base_names] + [ colname + '_25Q' for colname in base_names] +
             [ colname + '_50Q' for colname in base_names] + [ colname + '_75Q' for colname in base_names] + 
             [ colname + '_100Q' for colname in base_names] )
    
    # mean
    F = F + window_data.mean().tolist()
    names = names + [ colname + '_mean' for colname in base_names]
    
    # std
    F = F + window_data.std().tolist()
    names = names + [ colname + '_std' for colname in base_names]
    
    return F, names
```
